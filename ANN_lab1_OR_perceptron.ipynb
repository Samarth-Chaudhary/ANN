# Lab 1: Perceptron Simulation using NumPy (OR Gate)

A single **Perceptron** implemented from scratch using only NumPy, trained on the logical **OR** gate using the Perceptron Learning Rule (online/stochastic updates).

```python
import numpy as np
#── Activation ──
def step(x):
    return np.where(x >= 0, 1, 0)
#── Perceptron (functional style) ──
def train_perceptron(X, y, lr=0.1, epochs=10):
    weights = np.zeros(X.shape[1])
    bias = 0.0
    history = []
    
    for epoch in range(epochs):
        total_error = 0
        for xi, target in zip(X, y):
            output = step(np.dot(xi, weights) + bias)
            error = target - output
            weights += lr * error * xi
            bias    += lr * error
            total_error += abs(error)
        history.append(total_error)
        
    return weights, bias, history
# ── Training Data: OR Gate ──
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([0, 1, 1, 1])
 #── Train ──
weights, bias, history = train_perceptron(X, y, lr=0.1, epochs=10)
 #── Results ──
print("=" * 40)
print("   PERCEPTRON SIMULATION (OR Gate)")
print("=" * 40)
print(f"\nLearned Weights : {weights}")
print(f"Learned Bias    : {bias:.4f}")
print(f"\nTraining Errors per Epoch: {history}")
print("-" * 40)
print("\nPredictions:")
for xi in X:
    result = step(np.dot(xi, weights) + bias)
    print(f"  {xi}  →  {int(result)}")
print("-" * 40)
```

## Observation / Interpretation

- The perceptron starts with **weights = [0, 0]** and **bias = 0**, so its first predictions are wrong for every input that should output `1`.
- The **error count drops from 2 → 2 → 1 → 0** across the first four epochs, and stays at **0** for the remaining six epochs — this shows the perceptron **converges** well before the 10 epoch budget is used up, confirming the OR gate is **linearly separable**.
- The final learned parameters are **weights = [0.1, 0.1]** and **bias = -0.1**. Geometrically, this defines the decision line `0.1·x1 + 0.1·x2 − 0.1 = 0`, i.e. `x1 + x2 = 1`. Any point on or above this line is classified `1`; only `(0,0)` falls strictly below it.
- The final predictions `[0,1,1,1]` exactly match the OR truth table, so the model has **100% training accuracy** after convergence.
- Because both weights ended up equal (`0.1`), the model treats `x1` and `x2` symmetrically — which makes sense, since OR itself is a symmetric function of its two inputs.
